```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Upload the file
uploaded = files.upload()

# Load the Mall Customers dataset
df = pd.read_csv('Mall_Customers.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
# Check for missing values
df.isnull().sum()

# Since there are no missing values, we can proceed
```

```python
# Convert 'Gender' column to numerical
df['Genre'] = df['Genre'].map({'Male': 0, 'Female': 1})

# Display the first few rows of the updated dataframe
df.head()
```

```python
# Display the summary statistics of the dataframe
df.describe()

# Display the structure of the dataframe
df.info()
```

```python
# Gender distribution
sns.countplot(x='Genre', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender (0 = Male, 1 = Female)')
plt.ylabel('Count')
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Annual income distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Annual Income (k$)'], kde=True, bins=30)
plt.title('Annual Income Distribution')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Spending score distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Spending Score (1-100)'], kde=True, bins=30)
plt.title('Spending Score Distribution')
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot
sns.pairplot(df)
plt.show()
```

```python
# Spending Score vs Annual Income
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Genre', data=df)
plt.title('Spending Score vs Annual Income')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
